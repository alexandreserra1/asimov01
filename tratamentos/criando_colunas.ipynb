```python
import pandas as pd
# dicionario com nome,sobrenome e idade atual
data = {'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo', 'Fábio', 'Giovana', 'Hugo', 'Isabela', 'João'],
        'sobrenome': ['Silva', 'Souza', 'Oliveira', 'Santos', 'Pereira', 'Ferreira', 'Gomes', 'Lima', 'Costa', 'Rodrigues'],
        'idade': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]}

df = pd.DataFrame(data)
df
```

```python
df['é gente boa'] = 'Sim'
df
```

```python
df['cidade que nasceu'] = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Brasília', 'Salvador', 'Fortaleza', 'Manaus', 'Curitiba', 'Recife', 'Porto Alegre']
df
```

```python
#metodo .insert
df.insert(2, 'cidade que mora', ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Brasília', 'Salvador', 'Fortaleza', 'Manaus', 'Curitiba', 'Recife', 'Porto Alegre'])
df
```

```python
# colunas com Series
df['Dobro de idade'] = df['idade'] * 2
df
```
